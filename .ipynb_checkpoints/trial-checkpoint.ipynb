```python
from helpers import make_integer, get_detections, get_iou


                
def get_color(tracks, dicto):
    i = 0
    j = 1
    
    while i != len(tracks) - 1:
        
        if j == len(tracks)-1:
            i += 1
            j = i + 1
        
        b1 = tracks[i]
        b2 = tracks[j]
        bbox1 = {'x1': b1[0], 'y1': b1[1],'x2': b1[2],'y2': b1[3]}
        bbox2 = {'x1': b2[0], 'y1': b2[1],'x2': b2[2],'y2': b2[3]}
        
        iou = get_iou(bbox1, bbox2)
        
        if iou > 0.0:
            if (b1[4],b2[4]) in dicto.keys():
                dicto[(b1[4],b2[4])] += 1
            else:
                dicto.update({(b1[4],b2[4]):1})
        else:
            if (b1[4],b2[4]) in dicto.keys():
                del dicto[(b1[4],b2[4])]
        
        j += 1
        
    
    return dicto
        
    
    
    
    
```

```python
import cv2
import numpy as np
import imutils
import time
# reading videos with imutils
from imutils.video import FileVideoStream
from imutils.video import FPS
import torch
from sort.sort import *

#create instance of SORT
mot_tracker = Sort() 

path1 = "PETS2009 1.avi"

# Loading yolov5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained = True)

model.eval()

model.conf = 0.50

model.classes = [0]

c = 0
dicto = {}
fm1 = []
fm5 = []
video = FileVideoStream(path1).start()
fps = FPS().start()
time.sleep(1.0)
fm =0
while video.more():
    frame = video.read()
    if frame is None:
        break

    
    preds = get_detections(model, frame)
    
    
    dets = preds.pred[0]
    
    #det = dets[0].unsqueeze(0)
    
    #print(det)
    # update SORT
    track_bbs_ids = mot_tracker.update(dets)
    
    dicto = get_color(track_bbs_ids, dicto)
    
    for tracks, fm_no in dicto.items():
        if fm_no >= 5:
            fm5.append(tracks[0])
            fm5.append(tracks[1])
            
        elif fm_no >= 1 and fm_no < 5:
            fm1.append(tracks[0])
            fm1.append(tracks[1])
            
        
            
    
    for box in track_bbs_ids:
        track_id = box[4]
        if track_id in fm1:
            color = (255,0,0)
            
        elif track_id in fm5:
            color = (0,0,255)
        else:
            color = (0,255,255)
            
        x1,y1,x2,y2 = box[:4]
        
        cv2.rectangle(frame, (x1,y1), (x2,y2), color, 1)
        cv2.putText(frame, str(int(track_id)),(x1+4, y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)
        
        
    
    
    
    c += 1
    if c==50:
        break
    
    
    #'''
            
            
        
    
    
    
    #display
    cv2.imshow('Video', frame)
    cv2.waitKey(1)
    fps.update()
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
   



# stop the timer and display FPS information
fps.stop()
print("[INFO] elasped time: {:.2f}".format(fps.elapsed()))
print("[INFO] approx. FPS: {:.2f}".format(fps.fps()))
video.stop()    
cv2.destroyAllWindows()
cv2.waitKey(1)
#'''
```
